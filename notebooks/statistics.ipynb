```python
import geopandas as gpd
import pandas as pd
```

```python
res = []
for city in [
    "Braga",
    "Camden",
    "Cugir",
    "Kozani",
    "Lambeth",
    "Milan_metropolitan",
    "Riga",
    "Westminster",
    "Zaragoza",
]:
    print(city)
    poly = gpd.read_file(f"../data/processed/1_cities_boundaries/{city}.gpkg")
    poly = poly.to_crs(poly.estimate_utm_crs())
    streets = gpd.read_file(f"../data/processed/{city}/{city}_graph_2_dense.gpkg", layer="edges")
    streets = streets.to_crs(streets.estimate_utm_crs())
    amenities = gpd.read_file(f"../data/processed/{city}/{city}_features_3_dense.gpkg")
    res.append(
        [
            city,
            round(poly.geometry[0].area / 1000000, 1),
            len(streets),
            round(streets.geometry.length.sum() / 1000),
            round(streets[streets["pedestrian_infrastructure"] == "True"].geometry.length.sum() / 1000),
            round(streets[streets["cycling_infrastructure"] == "True"].geometry.length.sum() / 1000),
            len(amenities[amenities["type"] == "shop"]),
            len(amenities[amenities["type"] == "crossing"]),
            len(amenities[amenities["type"] == "traffic_signals"]),
        ]
    )
df = pd.DataFrame(
    data=res,
    columns=[
        "Cityname",
        "Area (km2)",
        "Number of edges",
        "Total road length (km)",
        "Pedestrian road length (km)",
        "Cycling road length (km)",
        "Number of shops",
        "Number of crossings",
        "Number of traffic signals",
    ]
)
```
